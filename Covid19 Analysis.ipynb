```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv("covid19_Confirmed_dataset.csv")
dataset.head()
```

```python
dataset.shape
```

```python
#deleting columns
df = dataset.drop(["Lat", "Long"], axis = 1, inplace = True)
```

```python
dataset.head()
```

```python
#aggregate the rows by country
corona_dataset_aggregated = dataset.groupby("Country/Region").sum()
```

```python
corona_dataset_aggregated.head()
```

```python
corona_dataset_aggregated.shape
```

```python
#visualise data related to countries
corona_dataset_aggregated.loc["China"].plot()
corona_dataset_aggregated.loc["India"].plot()
corona_dataset_aggregated.loc["Spain"].plot()
#Tells which line is which
plt.legend()
```

```python
corona_dataset_aggregated.loc["China"].plot()
```

```python
#checking for first three days
corona_dataset_aggregated.loc["China"][:3].plot()
```

```python
#calculate the first derivative of the curve
corona_dataset_aggregated.loc["China"].diff().plot()
```

```python
#maximum infection rate
corona_dataset_aggregated.loc["China"].diff().max()
```

```python
corona_dataset_aggregated.loc["India"].diff().max()
```

```python
#list of maximum rate for all countries stored in as list
countries = list(corona_dataset_aggregated.index)
max_infection_rates = []

for c in countries:
    max_infection_rates.append(corona_dataset_aggregated.loc[c].diff().max())

corona_dataset_aggregated["Maximum infection rate"] = max_infection_rates
```

```python
corona_dataset_aggregated.head()
```

```python
#create a new dataframe
corona_data = pd.DataFrame(corona_dataset_aggregated["Maximum infection rate"])
```

```python
corona_data
```
